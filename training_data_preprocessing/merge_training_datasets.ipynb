```python
top_features = ['patient_id', 'num_care_events', 'last_care_dt', 'num_care_gaps', 'has_next_care', 'has_last_care', 'type_MEDICATION ADHERENCE', 'type_SCREENING',
       'subtype_BREAST CANCER', 'subtype_CHOLESTEROL',
       'subtype_COLORECTAL CANCER', 'subtype_DIABETES',
        'subtype_HYPERTENSION', 'count_msrmnt_type', 'care_duration_days', 'msrmnt_type_diversity',
       'msrmnt_subtype_diversity',
       'days_until_next_care', 'weighted_care_score']
```

```python
diagnosis=pd.read_csv('Dataset/updated_diagnosis_risk.csv')
```

```python
diagnosis.columns
```

```python
top_features = ['patient_id', 'num_chronic_conditions',
       'cond_CANCER', 'has_cancer_history','diabetes_count',
       'chronic_condition_ratio', 'cancer_count', 'diabetes_count',
       'hypertension_count']
```

```python
diagnosis= diagnosis[top_features]
```

```python
patient=pd.read_csv('Dataset/updated_patient_risk.csv')
```

```python
patient.columns
```

```python
top_features = [
    'patient_id',
    'hotspot_recency_norm',
    'age',
    'hot_spotter_chronic_flag',
    'chronic_x_readmission',
    'hot_spotter_readmission_flag',
    'elderly_flag'
]
```

```python
patient=patient[top_features]
```

```python
visit=pd.read_csv('Dataset/updated_visit_risk.csv')
```

```python
visit.columns
```

```python
top_features = [
    'patient_id',
    'num_inpatient_visits',
    'num_follow_ups',
    'inpatient_visit_ratio',
    'num_visits',
    'followup_ratio',
    'num_er_visits',
    'days_between_first_last_visit',
    'er_visit_ratio',
    'visit_frequency_per_month'
]
```

```python
visit=visit[top_features]
```

```python
risk=pd.read_csv('Dataset/risk.csv')
risk.columns
```

```python
print(patient.shape)
print(visit.shape)
print(diagnosis.shape)
print(care.shape)
```

```python
merged_df = (
    patient
    .merge(visit, on="patient_id", how="left")
    .merge(diagnosis, on="patient_id", how="left")
    .merge(care, on="patient_id", how="left")
)

merged_df = merged_df.merge(risk, on="patient_id", how="left")

print(merged_df.shape)
# print(merged_df.head())
```

```python
merged_df_clean = merged_df.copy()


for col in merged_df_clean.select_dtypes(include=["number"]).columns:
    merged_df_clean[col] = merged_df_clean[col].fillna(0)


for col in merged_df_clean.select_dtypes(include=["object", "category"]).columns:
    merged_df_clean[col] = merged_df_clean[col].fillna("Unknown")
```

```python
merged_df_clean.to_csv("merged_training_dataset.csv", index=False)
```
