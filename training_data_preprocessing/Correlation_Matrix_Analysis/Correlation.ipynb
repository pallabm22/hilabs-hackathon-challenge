```python
risk=pd.read_csv('../Dataset/risk.csv')
risk.columns
```

```python
print(patient.shape)
print(visit.shape)
print(diagnosis.shape)
print(care.shape)
```

```python
print(patient_all.shape)
print(visit_all.shape)
print(diagnosis_all.shape)
print(care_all.shape)
```

```python
merged_df_all = (
    patient_all
    .merge(visit_all, on="patient_id", how="left")
    .merge(diagnosis_all, on="patient_id", how="left")
    .merge(care_all, on="patient_id", how="left")
)

merged_df_all = merged_df_all.merge(risk, on="patient_id", how="left")

print(merged_df_all.shape)
# print(merged_df.head())
```

```python
merged_df = (
    patient
    .merge(visit, on="patient_id", how="left")
    .merge(diagnosis, on="patient_id", how="left")
    .merge(care, on="patient_id", how="left")
)

merged_df = merged_df.merge(risk, on="patient_id", how="left")

print(merged_df.shape)
# print(merged_df.head())
```

```python
merged_df_clean_all = merged_df_all.copy()


for col in merged_df_clean_all.select_dtypes(include=["number"]).columns:
    merged_df_clean_all[col] = merged_df_clean_all[col].fillna(0)

for col in merged_df_clean_all.select_dtypes(include=["object", "category"]).columns:
    merged_df_clean_all[col] = merged_df_clean_all[col].fillna("Unknown")
```

```python
merged_df_clean = merged_df.copy()


for col in merged_df_clean.select_dtypes(include=["number"]).columns:
    merged_df_clean[col] = merged_df_clean[col].fillna(0)

for col in merged_df_clean.select_dtypes(include=["object", "category"]).columns:
    merged_df_clean[col] = merged_df_clean[col].fillna("Unknown")
```

```python
numeric_df = merged_df_clean.select_dtypes(include=[float, int])

corr_matrix = numeric_df.corr()
risk_corr = corr_matrix['risk_score'].sort_values(ascending=False)
print(risk_corr)
```

```python
plt.figure(figsize=(15,12))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title("Full Correlation Matrix")
plt.show()
```

```python
numeric_df_all = merged_df_clean_all.select_dtypes(include=[float, int])

corr_matrix_all = numeric_df_all.corr()
risk_corr_all = corr_matrix_all['risk_score'].sort_values(ascending=False)

print(risk_corr_all)
```

```python
plt.figure(figsize=(15,12))
sns.heatmap(corr_matrix_all, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title("Full Correlation Matrix")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(24, 10))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    cbar=True,
    ax=axes[0]
)
axes[0].set_title("Correlation Matrix (Filtered)", fontsize=14)
sns.heatmap(
    corr_matrix_all,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    cbar=True,
    ax=axes[1]
)
axes[1].set_title("Correlation Matrix (All Features)", fontsize=14)

plt.tight_layout()
plt.savefig('correlation.png', dpi=300)
plt.show()
```
